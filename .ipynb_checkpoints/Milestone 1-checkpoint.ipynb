````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df22 = pd.read_csv("kaggle_survey_2022_responses.csv")
df21 = pd.read_csv("kaggle_survey_2021_responses.csv")
df20 = pd.read_csv("kaggle_survey_2020_responses.csv")
````

Q11:For how many years have you been writing code and/or programming?  
Q23:Select the title most similar to your current role (or most recent title if retired): - Selected Cho...  
Q24: In what industry is your current employer/contract (or your most recent employer if retired)? - Sele.  
Q25:What is the size of the company where you are employed  
Q29:What is your current yearly compensation (approximate $USD)?  

````python
df22 = df22.rename(columns={'Duration (in seconds)': 'Duration','Q2':'Age','Q3':'Gender','Q4':'Country','Q8':'Education',
                   'Q23':'Occupation','Q24':'Industry','Q25':'Comp_size','Q29':'Salary'})
df21 = df21.rename(columns={'Time from Start to Finish (seconds)': 'Duration','Q1':'Age','Q2':'Gender','Q3':'Country','Q4':'Education',
                   'Q5':'Occupation','Q20':'Industry','Q21':'Comp_size','Q25':'Salary'})
df20 = df20.rename(columns={'Time from Start to Finish (seconds)': 'Duration','Q1':'Age','Q2':'Gender','Q3':'Country','Q4':'Education',
                   'Q5':'Occupation','Q20':'Comp_size','Q24':'Salary'})
# select useful columns to analysis and drop other columns
df22 = df22[['Duration','Age','Gender','Country','Education','Occupation','Salary','Comp_size','Industry']].drop(index=0)
df21 = df21[['Duration','Age','Gender','Country','Education','Occupation','Salary','Comp_size','Industry']].drop(index=0)
df20 = df20[['Duration','Age','Gender','Country','Education','Occupation','Salary','Comp_size']].drop(index=0)
````

````python
# drop records with NaN
df22.dropna(inplace=True)
df22.reset_index(inplace=True)

df21.dropna(inplace=True)
df21.reset_index(inplace=True)

df20.dropna(inplace=True)
df20.reset_index(inplace=True)
````

# Deal with Salary
## 1 . del $

````python
# 1. del $
df22['Salary'] = df22.loc[:,'Salary'].str.replace('$', '')
df21['Salary'] = df21.loc[:,'Salary'].str.replace('$', '')
df20['Salary'] = df20.loc[:,'Salary'].str.replace('$', '')
````

## 2. keep the salary range same
**min**  
22和21相同，20是把500,000和>1,000,000统一变为了>500,000 ; 将22和21变为20的形式

min = array(['0', '1,000', '10,000', '100,000', '125,000', '15,000', '150,000',
       '2,000', '20,000', '200,000', '25,000', '250,000', '3,000',
       '30,000', '300,000', '4,000', '40,000', '5,000', '50,000',
       '60,000', '7,500', '70,000', '80,000', '90,000', '> 500,000']
       
       
**max**   
22和21相同，20的500,000应该对应21和22年的499,999  
22和21年的999,999 和none都应该变成 none


````python
# First Step : keep the salary_range the same for the 3 years.
# the max range is >500,000  so change 2022 and 2011 . 500,000-999,999 and >1,000,000  --> >500,000
df22['Salary'] = df22['Salary'].apply(lambda x: '>500,000' if x =='500,000-999,999' or x == '>1,000,000' else x)
df21['Salary'] = df21['Salary'].apply(lambda x: '>500,000' if x =='500,000-999,999' or x == '>1,000,000' else x)

# change the 1. 2020 300,000-500,000 --> 300,000-499,999 (2022,2021)
df20['Salary'] = df20['Salary'].apply(lambda x: '300,000-499,999' if x =='300,000-500,000' else x)

# adjust 2020 > 500,00 --> >500,000
df20['Salary'] = df20['Salary'].apply(lambda x: '>500,000' if x =='> 500,000' else x)
````

````python
df22[['min_Salary', 'max_Salary']] = df22['Salary'].str.split('-', expand=True)
df21[['min_Salary', 'max_Salary']] = df21['Salary'].str.split('-', expand=True)
df20[['min_Salary', 'max_Salary']] = df20['Salary'].str.split('-', expand=True)

df22 = df22.sort_values(by='min_Salary', ascending=True)
df21 = df21.sort_values(by='min_Salary', ascending=True)
df20 = df20.sort_values(by='min_Salary', ascending=True)
````
